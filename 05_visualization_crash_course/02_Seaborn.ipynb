```python
# https://seaborn.pydata.org
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df = pd.read_csv('heart.csv')
```

```python
df.head()
```

```python
# distribution plot/ histogram
sns.distplot(df['age'])
```

```python
sns.distplot(df['age'], kde=False)
```

```python
sns.distplot(df['age'], kde=False, bins=2)
```

```python
# change the size of the plot
plt.figure(figsize=(8,4))
sns.distplot(df['age'], kde=False, color='red')
```

```python
df.head()
```

```python
# group by sex. In ML display how balance are data
sns.countplot(x='sex', data=df)
```

```python
sns.countplot(x='cp',data=df,hue='sex')
```

```python
sns.countplot(x='cp',data=df,palette='terrain')
```

```python
# ------- Box Plot -----------
```

```python
sns.boxplot(x='sex', y='age', data=df)
```

```python
sns.boxplot(x='target', y='thalach', data=df)
```

```python
sns.boxplot(x='target', y='thalach', data=df, hue='sex')
```

```python
# ----- scatter lot -------------
sns.scatterplot(x='chol', y='trestbps',data=df)
```

```python
sns.scatterplot(x='chol',y='trestbps',data=df,hue='sex',palette='Dark2')
```

```python
sns.scatterplot(x='chol',y='trestbps',
            data=df,hue='sex',palette='Dark2',size='age')
```
