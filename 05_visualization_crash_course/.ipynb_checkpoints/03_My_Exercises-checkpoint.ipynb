```python
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
diamonds = pd.read_csv('diamonds.csv')
```

```python
diamonds.head()
```

```python
diamonds.info()
```

```python
# TASK: Create a scatterplot of price versus carat
sns.scatterplot(x='carat',y='price',data=diamonds)
```

```python
plt.xlim(0,6)
plt.ylim(0,20000)
sns.scatterplot(x='carat',y='price',data=diamonds,alpha=0.4, edgecolor="none")
```

```python
#help(sns.scatterplot)
```

```python
plt.figure(figsize=(12,10))
plt.xlim(0,6)
plt.ylim(0,20000)
sns.scatterplot(x='carat',y='price',data=diamonds,alpha=0.4, edgecolor="none")
```

```python
plt.figure(figsize=(8,4))
sns.distplot(diamonds['price'], kde=False)
```

```python
diamonds.head()
```

```python
sns.countplot(x='cut',data=diamonds)
```

```python
sns.boxplot(x='cut',y='price',data=diamonds)
```

```python
help(sns.boxplot)
```

```python
sns.boxplot(x='cut',y='price',data=diamonds, palette="Blues",
            order=["Fair","Ideal","Good","Very Good","Premium"])       
```

```python
my_pal = {cut: "red" if cut == "Good" else "b" for cut in diamonds.cut.unique()}
sns.boxplot(x='cut',y='price',data=diamonds, palette=my_pal,
            order=["Fair","Ideal","Good","Very Good","Premium"])  
```
